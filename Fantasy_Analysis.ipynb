```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
link = "https://raw.githubusercontent.com/carsonbennett1/Hockey-Player-Analysis-Project/refs/heads/main/stats.csv"

dataSet = pd.read_csv(link)
df = pd.DataFrame(dataSet)
df
```

#### Fantasy Player Nationality at Week 1

```python
#Get Pie Chart Labels
labels = ['Canadian', 'American', 'Finish', 'Swiss', 'Swedish']

# Get Player Nationalities
# Players must have played in Week6
canadian = df[(df['Nationality'] == 'CAN') & (~pd.isna(df['Week13']))]
american = df[(df['Nationality'] == 'USA') & (~pd.isna(df['Week13']))]
swiss = df[(df['Nationality'] == 'SWI') & (~pd.isna(df['Week13']))]
finish = df[(df['Nationality'] == 'FIN') & (~pd.isna(df['Week13']))]
swedish = df[(df['Nationality'] == 'SWE') & (~pd.isna(df['Week13']))]

# Get Player Nationality Numbers
nums = [canadian['Name'].count(), american['Name'].count(), finish['Name'].count(), 
        swiss['Name'].count(), swedish['Name'].count()]

plt.figure(figsize=(10,6))

# Show Distribution
plt.pie(nums, labels=labels, autopct='%1.1f%%', startangle=90,
    colors=['#ff0c02', '#0536bc', '#0091ff', '#ff301a', '#eae500'])
plt.title("Player Nationality at Week 13")
plt.show()
```

#### Centremen Fantasy Week Points Totals Compared

```python
mcdavid_df = df[df['Name'] == 'C.McDavid']
weeks = [col for col in df.columns if 'Week' in col]
mcdavid_points = mcdavid_df[weeks].values.flatten()
hischier_df = df[df['Name'] == 'N.Hischier']
point_df = df[df['Name'] == 'B.Point']
scheifele_df = df[df['Name'] == 'M.Scheifele']

hischier_points = hischier_df[weeks].values.flatten()
point_points = point_df[weeks].values.flatten()
scheifele_points = scheifele_df[weeks].values.flatten()

mcdavid_points = np.nan_to_num(mcdavid_points, nan=0)
hischier_points = np.nan_to_num(hischier_points, nan=0)
point_points = np.nan_to_num(point_points, nan=0)
scheifele_points = np.nan_to_num(scheifele_points, nan=0)


plt.figure(figsize=(12,7))

plt.title("Centremen Fantasy Week Points Totals Compared")
plt.xlabel("Weeks")
plt.ylabel("Fantasy Week Total")

plt.plot(weeks,hischier_points,color='#d10003')
plt.plot(weeks,point_points,color='#1b00d1')
plt.plot(weeks,scheifele_points,color='#0d0064')
plt.plot(weeks,mcdavid_points,color='orange')
plt.legend(["Hischier","Point","Scheifele","McDavid"], loc="upper right")

plt.grid(color='green', linestyle='--', linewidth='0.5')

plt.show()
```

#### Position Point Analysis

```python
centremen = df[df['Position'] == 'C']
winger = df[df['Position'] == 'LW/RW']
left_winger = df[df['Position'] == 'LW']
right_winger = df[df['Position'] == 'RW']
centre_right = df[df['Position'] == 'C/RW']
centre_left = df[df['Position'] == 'C/LW']


centre_points = centremen[weeks].values.flatten()
centre_right_points = centre_right[weeks].values.flatten()
centre_left_points = centre_left[weeks].values.flatten()
winger_points = winger[weeks].values.flatten()
left_winger_points = left_winger[weeks].values.flatten()
right_winger_points = right_winger[weeks].values.flatten()
defensemen_points = defensemen[weeks].values.flatten()
goalies_points = goalies[weeks].values.flatten()


centre_points = np.nan_to_num(centre_points, nan=0)
centre_right_points = np.nan_to_num(centre_right_points, nan=0)
centre_left_points = np.nan_to_num(centre_left_points, nan=0)
winger_points = np.nan_to_num(winger_points, nan=0)
left_winger_points = np.nan_to_num(left_winger_points, nan=0)
right_winger_points = np.nan_to_num(right_winger_points, nan=0)
defensemen_points = np.nan_to_num(defensemen_points, nan=0)
goalies_points = np.nan_to_num(goalies_points, nan=0)


centre_sum = round(centre_points.sum(),1)
centre_right_sum = round(centre_right_points.sum(),1)
centre_left_sum = round(centre_left_points.sum(),1)
winger_sum = round(winger_points.sum(),1)
left_winger_sum = round(left_winger_points.sum(),1)
right_winger_sum = round(right_winger_points.sum(),1)
defensemen_sum = round(defensemen_points.sum(),1)
goalie_sum = round(goalies_points.sum(),1)

labels = ['C', 'C/RW', 'RW', 'C/LW', 'LW', 'LW/RW', 'D', 'G']

nums = [centre_sum, centre_right_sum, right_winger_sum, centre_left_sum, left_winger_sum,
        winger_sum, defensemen_sum, goalie_sum]

explode = (0,0,0,0,0,0,0.1,0.1)

plt.figure(figsize=(9,6))

plt.pie(nums, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90,
    colors=['#009aff', '#0065a6', '#d8f700', '#00c99a', '#8d00c8', '#8c67ff', '#02e100', '#e70800'])
plt.title('Player Position Contribution to Total Points')
plt.show()

```

#### Position Point Totals

```python
x_ticks = np.arange(0,1500,100)

plt.figure(figsize=(9,7))

h_bar_graph = plt.barh(labels, nums)
plt.bar_label(h_bar_graph,labels=nums,label_type='edge', padding=5)

h_bar_graph[0].set_color('#009aff')
h_bar_graph[1].set_color('#0065a6')
h_bar_graph[2].set_color('#d8f700')
h_bar_graph[3].set_color('#00c99a')
h_bar_graph[4].set_color('#8d00c8')
h_bar_graph[5].set_color('#8c67ff')
h_bar_graph[6].set_color('#02e100')
h_bar_graph[7].set_color('#e70800')

plt.xticks(x_ticks)
plt.title('Player Position Contribution to Total Points')
plt.ylabel('Position')
plt.xlabel('Total Fantasy Points')
plt.show()
```
