```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
link = "https://raw.githubusercontent.com/carsonbennett1/Hockey-Player-Analysis-Project/refs/heads/main/stats.csv"

dataSet = pd.read_csv(link)
df = pd.DataFrame(dataSet)
df
```

#### Fantasy Player Nationality at Week 1

```python
#Get Pie Chart Labels
labels = ['Canadian', 'American', 'Finish', 'German', 'Swiss', 'Swedish']

# Get Player Nationalities
# Players must have played in Week6
canadian = df[(df['Nationality'] == 'CAN') & (~pd.isna(df['Week1']))]
american = df[(df['Nationality'] == 'USA') & (~pd.isna(df['Week1']))]
german = df[(df['Nationality'] == 'GER') & (~pd.isna(df['Week1']))]
swiss = df[(df['Nationality'] == 'SWI') & (~pd.isna(df['Week1']))]
finish = df[(df['Nationality'] == 'FIN') & (~pd.isna(df['Week1']))]
swedish = df[(df['Nationality'] == 'SWE') & (~pd.isna(df['Week1']))]

# Get Player Nationality Numbers
nums = [canadian['Name'].count(), american['Name'].count(), finish['Name'].count(), german['Name'].count(), 
        swiss['Name'].count(), swedish['Name'].count()]

plt.figure(figsize=(10,6))

# Show Distribution
plt.pie(nums, labels=labels, autopct='%1.1f%%', startangle=90,
    colors=['#f10707', '#004ec0', '#7daef7', '#ffd131', '#a50002', '#f1db00'])
plt.title("Player Nationality at Week 1")
plt.show()
```

#### Fantasy Player Nationality Currently (Week13)

```python
#Get Pie Chart Labels
labels = ['Canadian', 'American', 'Finish', 'Swiss', 'Swedish']

# Get Player Nationalities
# Players must have played in Week6
canadian = df[(df['Nationality'] == 'CAN') & (~pd.isna(df['Week13']))]
american = df[(df['Nationality'] == 'USA') & (~pd.isna(df['Week13']))]
swiss = df[(df['Nationality'] == 'SWI') & (~pd.isna(df['Week13']))]
finish = df[(df['Nationality'] == 'FIN') & (~pd.isna(df['Week13']))]
swedish = df[(df['Nationality'] == 'SWE') & (~pd.isna(df['Week13']))]

# Get Player Nationality Numbers
nums = [canadian['Name'].count(), american['Name'].count(), finish['Name'].count(), 
        swiss['Name'].count(), swedish['Name'].count()]

plt.figure(figsize=(10,6))

# Show Distribution
plt.pie(nums, labels=labels, autopct='%1.1f%%', startangle=90,
    colors=['#f10707', '#004ec0', '#ffd131', '#a50002', '#f1db00'])
plt.title("Player Nationality at Week 13")
plt.show()
```

#### Centremen Fantasy Week Points Totals Compared

```python
mcdavid_df = df[df['Name'] == 'C.McDavid']
weeks = [col for col in df.columns if 'Week' in col]
mcdavid_points = mcdavid_df[weeks].values.flatten()
hischier_df = df[df['Name'] == 'N.Hischier']
point_df = df[df['Name'] == 'B.Point']
scheifele_df = df[df['Name'] == 'M.Scheifele']

hischier_points = hischier_df[weeks].values.flatten()
point_points = point_df[weeks].values.flatten()
scheifele_points = scheifele_df[weeks].values.flatten()


plt.figure(figsize=(12,7))

plt.title("Centremen Fantasy Week Points Totals Compared")
plt.xlabel("Weeks")
plt.ylabel("Fantasy Week Total")

plt.plot(weeks,hischier_points,color='#d10003')
plt.plot(weeks,point_points,color='#1b00d1')
plt.plot(weeks,scheifele_points,color='#0d0064')
plt.plot(weeks,mcdavid_points,color='orange')
plt.legend(["Hischier","Point","Scheifele","McDavid"], loc="upper right")

plt.grid(color='green', linestyle='--', linewidth='0.5')

plt.show()
```

#### Centre Point Summary

```python
mcdavid_points = np.nan_to_num(mcdavid_points, nan=0)
hischier_points = np.nan_to_num(hischier_points, nan=0)
point_points = np.nan_to_num(point_points, nan=0)
scheifele_points = np.nan_to_num(scheifele_points, nan=0)

x = ['C.McDavid', 'B.Point', 'N.Hischier', 'M.Scheifele']
y = [mcdavid_points.sum(), point_points.sum(), hischier_points.sum(), scheifele_points.sum()]


bar_graph = plt.bar(x,y)
bar_graph[0].set_color('orange')
bar_graph[1].set_color('#1b00d1')
bar_graph[2].set_color('#d10003')
bar_graph[3].set_color('#0d0064')

plt.title('Centremen Point Totals')
plt.ylabel('Total Fantasy Points')
plt.xlabel('Centremen')
plt.grid(axis = 'y')
plt.show()


```
